```python
# missing values
air_quality_data.isnull().sum()
# it shows that there are no null values in this dataset
# but Missing values are tagged with -200 value.---mentioned in the UCI repository
```

```python
# counting the no. of times -200 appears in the data
air_quality_data.isin([-200]).sum(axis=0)
```

```python
# Now comes handling the missing values
# first convert the -200 to NAN and
# then convert the NAN values to the mean of the particular column
air_quality_data = air_quality_data.replace(to_replace = -200, value = np.NAN)
```

```python
# now checking the missing values(NAN) in the dataframe
air_quality_data.isnull().sum()
```

```python
# now calculating the mean if the dataframe
air_quality_data.select_dtypes(include=np.number).mean()
```

```python
# convert the NAN values to the mean of the particular column
```

```python
air_quality_data = air_quality_data.fillna(air_quality_data.select_dtypes(include=np.number).mean())
```

# FB Prophet Model

```python
# converting the date from DD/MM/YYYY to YYYY-MM-DD
air_quality_data['Date'] = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
```

```python
air_quality_data.head()
```

```python
# covert the time from . to the :
air_quality_data['Time'] = pd.to_datetime(air_quality_data['Time'], format='%H.%M.%S').dt.time
```

```python
air_quality_data.head()
```

```python
print(type(air_quality_data['Time']))
```

```python
# combining 2 series to a pandas dataframe
date_time = pd.concat([air_quality_data['Date'], air_quality_data['Time']], axis=1)
```

```python
date_time.head()
```

```python
# combining these two columns
date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

```python
# creating a new dataframe
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y'] = air_quality_data['T']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
# training the prophet model
model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
data.tail()
```

```python
# Now we will forcast the RH
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
